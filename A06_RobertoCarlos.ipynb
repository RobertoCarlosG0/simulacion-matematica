Resultado final:

A mi entendimiento del problema, no me interesa ni el tiempo de envío de los materiales de los aretes y anillos, ni el tiempo de envío de los materiales de los collares; solo me interesa el envío que tardó menos. Ya que, si suponemos que Raúl no puede hacer dos tipos de accesorios al mismo tiempo (en un mismo día), e hizo todos los pedidos de materiales el mismo día, tendríamos un escenario donde a Raúl le llegaron primero los materiales de las pulseras y se puso a trabajar haciéndolas durante 50 días, tiempo suficiente para que le llegaran los otros pedidos. Por eso es que digo que solo interesa el tiempo que tarda el envío más corto en este caso.

Teniendo esos parámetros, los días que tardaría Raúl en tener todo listo serían:

2.5 dias de envio pulseras +
50 dias de hacer pulseras +
20 dias en hacer aretes +
19 dias en hacer anillos +
48 dias de hacer collares

Lo que daría como resultado 139 días y medio, que serían aproximadamente 4 meses y medio.

Por el contrario si Raúl pudiera hacer los accesorios al mismo tiempo pues simplemente nos importaria el tiempo de proceso mas largo que en este caso seria los collares que tardan aproximadamente 17.5 dias en llegarle a Raul mas los 48 dias que le toma en promedio hacerlos

```python
def simR1(media, std, start, dias, N) :
    sim_R1 = np.random.normal(media, std, size=(dias, N)) +1
    fechas = pd.date_range(start, periods= dias)
    return pd.pandas.DataFrame(sim_R1, index=fechas)
```
